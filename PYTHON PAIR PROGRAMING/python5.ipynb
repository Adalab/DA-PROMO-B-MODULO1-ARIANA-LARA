```python
# Variables predefinidas
hola= '¡¡Bienvenidos al 3 en raya!!'
adios='El juego ha terminado, ¡¡gracias por jugar!!'

# Inicio del juego
print(hola)

# Definir el tablero con todo "-"
tablero = [["-", "-", "-"], ["-", "-", "-"], ["-", "-", "-"]]

# Imprimir tablero
print(tablero[0])  
print(tablero[1])
print(tablero[2])

# Input de la usuaria
q=input('símbolo elegido     posición filas, posición columnas')
print('La usuaria 1 eligió la variable y la casilla:', q)

# Obtenemos los datos que necesitamos a partir del input
input_usuaria1= q.split()
num1= input_usuaria1[0]
coor=input_usuaria1[1]

coor = tuple(int(num1)-1 for num1 in coor.split(","))

fila= coor[0]
columna= coor[1]

#Le indicamos al programa donde debe poner la variable y actualizamos tablero
tablero[fila][columna] = num1

print(tablero[0])  
print(tablero[1])
print(tablero[2])

#Repetimos todo el proceso para la usuaria 2
q2=input('símbolo elegido     posición filas, posición columnas')
print('La usuaria 2 eligió la variable y la casilla:', q2)

input_usuaria2= q2.split()
num2= input_usuaria2[0]
coor=input_usuaria2[1]

coor = tuple(int(num2)-1 for num2 in coor.split(","))

fila= coor[0]
columna= coor[1]

tablero[fila][columna] = num2

print(tablero[0])
print(tablero[1])
print(tablero[2])

# Terminamos el juego
print('Los 2 jugadores que han participado son',num1, 'y', num2)
print(adios)























```
